Regular Expressions( RegEx)
---------------------------
--> This regualr expression or RegEx is a sequence of characters that forms a searching pattern.
--> To use this, we have to import re, which will unlock  the package

Basic Regex Functions in Python
--------------------------------
Function	    Meaning
1. re.match()	    Checks pattern at beginning
2. re.search()	    Finds first match anywhere
3. re.findall()	    Returns all matches
4. re.sub()	        Replaces matched text
5. re.split()	    Splits string

1. findall()
-------------
--> By using this function, it will find all sequence in the string
-->syntax
    re.findall("metachar", variable_name)

```python
#findall()
import re
any = "Python is a programming language"
so = re.findall("a", any)
print(so)
```

2. search()
------------
--> By using this function, it will only find first sequence in the string
--> Syntax
    re.search("metachar", variable_name)

```python
#search()
import re
any = "Python is a programming language"
so = re.search("a", any)
print(so)
```

Metacharacters
---------------
--> Metacharacters are used to form searching patterns

Important Special Characters
-----------------------------
Symbol	Meaning
.	    Any one character
^	    Starts with
$	    Ends with
*	    0 or more times
+	    1 or more times
?	    0 or 1 time
[]	    Set of characters
\d	    Digit
\w	    Letter/number/_
\s	    Space

1. []
--> In this meta character we can search for any character range specified in "[]" (square brackets)
--> Syntax
    re.findall("[specify range here]", variable_name)

```python
# Meta character - []
#Example-1 for findall()
import re
any = "This method can read entire file and return into list with"
so = re.findall("[a-z]", any)
print(so)
```

```python
# Meta character - []
#Example-2 for search()
import re
any = "By this function it will only find first sequence in the string"
so = re.search("[f]", any)
print(so)
```

2. .
--> This meta character will form from a searching pattern as it will take any single character usind (.)
--> Syntax
    re.findall("sequence", variable_name)

```python
#Meta character - .
#Example-1 Using findall()
import re
we = "hello"
the = re.findall("h...o", we)
print(the)
```

```python
#Meta character - .
#Example-1 Using findall()
import re
we = "hello"
the = re.search("h...o", we)
print(the)
```

3. ^
------
--> This is used to find the string starting with the sequence
--> Syntax
    re.findall("^sequence", variable name)

```python
#Metacharacter - ^
#Example-1 using findall()
import re
how = "This is used to find the string is starting with the sequence"
who = re.findall("^This is", how)
print(who)
```

```python
#Metacharacter - ^
#Example-2 using search()
import re
how = "This is used to find the string is starting with the sequence"
who = re.search("^This is", how)
print(who)
```

4. $
------
--> This is used to find the string ending with the sequeance or not
--> Syntax
re.findall("$", variable_name)

```python
#Metacharacter - $
#Example-1 using findall()
import re
out = "This is used to find the string is ending with the sequence"
one = re.findall("sequence$", out)
print(one)
```

```python
#Metacharacter - $
#Example-2 using search()
import re
out = "This is used to find the string is ending with the sequence"
one = re.search("sequence$", out)
print(one)
```

5. *
-----
--> This meta character will form a searching pattern as it will take any zero or more characters for (*)
--> Syntax
    re.search(".*", variable_name)

```python
#Metacharacter - *
#Example-1 using findall()
import re
roy = "This meta character will form a searching pattern as it will take any zero or more characters for (*)"
lr = re.findall("c.*r", roy)
print(lr)
```

```python
#Metacharacter - *
#Example-2 using search()
import re
roy = "This meta character will form a searching pattern as it will take any zero or more characters for (*)"
lr = re.search("T.*r", roy)
print(lr)
```

5. +
-----
--> This meta character will form a searching pattern as it will take any one or more characters for (*)
--> Syntax
    re.search(".*", variable_name)

```python
#Metacharacter - +
#Example-1 using findall()
import re
roy = "This meta character will form a searching pattern as it will take any one or more characters for (+)"
lr = re.findall("me.+a", roy)
print(lr)
```

```python
#Metacharacter - +
#Example-2 using search()
import re
roy = "This meta character will form a searching pattern as it will take any one or more characters for (+)"
lr = re.search("T.+", roy)
print(lr)
```
