6. ?
-----
This meta character will form a searching pattern as it will take any zero or one character for (?)
--> Syntax
    re.findall("char.?", variable_name)

```python
#Meta Character - ?
#Example 1 using findall()
any = "This meta character"
an = re.findall("Th.?", any)
print(an)
```

```python
#Meta Character - ?
#Example 2 using search()
any = "This meta character"
an = re.search("Th.?", any)
print(an)
```

8. {}
-------
--> This meta character will form a searching pattern as we can mention the size in the {}
--> Syntax
    re.search(".{size}", variable)

```python
#Meta Character - {}
#Example 1 using findall()
any = "This meta character will form a searching as it will take n number of characters"
am = re.findall(".{25}", any)
print(am)
```

```python
#Meta Character - {}
#Example 2 using search()
any = "This meta character will form a searching as it will take n number of characters"
am = re.search(".{25}", any)
print(am)
```

9. |
-----
This meta character will form a searching pattern as it considers right or left any string is present or not for (|)

```python
#Meta Character - |
#Example 1 using findall()
any = "This meta character will form"
an = re.findall("that|will", any)
print(an)
```

```python
#Meta Character - |
#Example 2 using search()
any = "This meta character will form"
an = re.search("that|will", any)
print(an)
```

Special Sequences in Python Regular Expressions
---------------------------------------------------
--> In Regular Expressions (Regex), special sequences are shortcuts that begin with a backslash \ and represent common character groups or positions.

--> Instead of writing long patterns, we use these short forms.

1. \A
----
Returns a match if the specified characters are at the beginning of the string

```python
#Special Sequence - \A
#Example using findall()
txt = "The rain in HYD"
#Chech if the string starts with "The":
x = re.findall(r"\AThe", txt)
print(x)
if x:
    print("Yes, there is a match")
else:
    print("No match")
```

2. \b
--> Returns a match where the specified characters are at the beginning or at the and end of a string

```python
#Special Sequence - \b
#Example using findall()
txt = "The rain in HYD"
#Chech if the string starts or ends with "HYD":
x = re.findall(r"\bHYD", txt)
print(x)
if x:
    print("Yes, there is atleast one match!")
else:
    print("No match")
```

2. \d
--> Returns a match where the string contains digits(numbers from 0-9)

```python
#Special Sequence - \d
#Example using findall()
txt = "The rain in 67 HYD"
#Chech if the string contains any digits (numbers from 0-9)
x = re.findall(r"\d", txt)
print(x)
if x:
    print("Yes, there is atleast one match!")
else:
    print("No match")
```

3. \D
--> Returns the match where the string does not contain digits

```python
#Special Sequence - \D
#Example using findall()
txt = "The rain in 67 HYD"
#Returns a match at every no-digit character
x = re.findall(r"\D", txt)
print(x)
```

3. \s
--> Returns a match where the string contains white space character

```python
#Special Sequence - \s
#Example using findall()
txt = "The rain in 67 HYD"
#Returns a match at every no-digit character
x = re.findall(r"\s", txt)
print(x)
if x:
    print("Yes, there is atleast one match!")
else:
    print("No match")
```

3. \S
--> Returns a match where the string DOES NOT contain a  white space character

```python
#Special Sequence - \S
#Example using findall()
txt = "The rain in 67 HYD"
#Returns a match at every no-digit character
x = re.findall(r"\S", txt)
print(x)
```
