Time and Date
---------------
%d --> Day
%m --> Month
%Y --> Year
%H --> Hours
%M --> Minutes
%S --> Seconds
%p --> AM/PM
%A --> Day name
%B --> Month name

```python
import datetime
```

```python
#Example
now = datetime.datetime.now()
print(now)
```

```python
today = datetime.date.today()
print(today.strftime("%d-%m-%Y"))
print(today.strftime("%A"))
print(today.strftime("%B"))
```
