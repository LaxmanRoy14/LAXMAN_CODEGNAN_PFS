 Polymorphism
 --------------
 --> This allows a object of different classes to be treated as instance of the same base class,
     with methods behaving differently based on the actual object type.

```python
#Example:
print(len("Python"))
print(len([1,2,3]))
```

Method Overloading
------------------
--> This defines multiple methods with the same name, but different parameters
    (number, type, or order) in the same class
--> Python does not support method overloading directly.
--> If multiple methods have same name, the latest one overrides previous methods.
--> We simulate overloading using default arguments or *args.

```python
#Example for Method Overloading
class calculator:
    def add(self, a, b, c=0):
        return a + b + c
obj = calculator()
print(obj.add(3,5))
print(obj.add(4,7,8))
```

```python
#Example for multiple args:
class calci:
    def add(self,*nums):
        return sum(nums)
obj = calci()
print(obj.add(2,3,5,7,9))
print(obj.add(1,345,67))
```

```python
#PRACTICE QUESTION FOR CALCULATOR
class Calculator:
    def multiply(self, *args):
        result = 1
        
        for num in args:
            result *= num
            
        print(result)


obj = Calculator()

obj.multiply(int(input("Enter First number: ")), int(input("Enter second number: ")))          # 6
obj.multiply(2, 3, 4)       # 24
obj.multiply(1, 2, 3, 4)    # 24
obj.multiply(5, 2, 3, 2)    # 60
```

Method Overriding
------------------
--> This occurs in the child class, redefining a parent class method with the signature for run time.

```python
#Example for Method Overriding
class animal:
    def speak(self):
        return "Sound"
class dog(animal):
    def speak(self):
        return"Woof"
obj = dog()
print(obj.speak())
```

```python
#Example2 for Method Overriding
class dad:
    def advice(self):
        return "Work hard"
class mom(dad):
    def advice(self):
        return "Take care"
obj = mom()
obj.advice()
```

Operator Overloading
---------------------
--> This customizes the operator like "+, -" for user-defined classes by implementing special methods..
    Example: __add__, __sub__

```python
class someone:
    def __init__(self, a, b):
        self.a = a
        self.b = b
    def __add__(self, other):
        return someone(self.a + other.a, self.b + other.b)
    def __str__(self):
        return f"{self.a}, {self.b}"
any = someone(2, 3)
so = someone(5, 9)
print(any + so)
```

Data Abstraction
-----------------
--> This hides complex implementation details, exposing only essential features via abstract class or interface

```python
from abc import ABC, abstractmethod
class shape(ABC):
    @abstractmethod
    def area(self):
        pass
class circle(shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14 * self.radius ** 2
Circle = circle(5)
print(Circle.area())
```
