Error Handling
---------------
try
→ Python tries to run the code inside it.

except
→ If an error happens, Python runs this block instead of stopping the program.

In One Line:
try = test the code
except = handle the error safely

```python
#Example
try:
    print(b)
except:
    print("Variable not found")
```

What happens here?
→ Python tries to print b
→ But b is not defined, so an error happens
→ Instead of crashing, except runs and prints: "Variable not found"

else Block in Python Exception Handling
-----------------------------------------
→ else is used with try and except.
→ The else block runs only if no error happens in the try block.
→ If an error happens, except runs and else is skipped.

Syntax:
try:
    # risky code
except:
    # runs if error occurs
else:
    # runs if no error

```python
#Example
try:
    a = 10
    b = 2
    print(a / b)
except:
    print("Error occurred")
else:
    print("Division successful")
```

finally Block in Python Exception Handling
-------------------------------------------
→ finally is used with try and except.
→ The finally block always runs, whether an error happens or not.
→ It is commonly used for cleanup work like closing files, database connections, releasing resources, etc.

Syntax
-------
try:
    # risky code
except:
    # runs if error occurs
finally:
    # always runs

```python
#Example (No error)
try:
    print(10 / 2)
except:
    print("Error")
finally:
    print("This always runs")
```

```python
#Example (with error)
try:
    print(10 / 0)
except:
    print("Cannot divide by zero")
finally:
    print("This always runs")
```

Simple Meaning
try → test code
except → handle error
else → run if no error
finally → run no matter what

```python
#Example for Error Handling
try:
    num_1 = int(input("Enter 1st number: "))
    num_2 = int(input("Enter 2nd number: "))
    result = num_1 / num_2
except ValueError:
    print("Please Enter a valid number.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
else:
    print(f"Result = {result}")
finally:
    print("Program is completed")
```

```python
try:
    balance = 5000

    amount = int(input("Enter withdrawal amount: "))

    if amount > balance:
        raise Exception("Insufficient Balance")

    balance = balance - amount

except ValueError:
    print("Please enter numbers only.")

except Exception as e:
    print(e)

else:
    print("Withdrawal successful")
    print("Remaining Balance =", balance)

finally:
    print("Thank you for using ATM")
```
