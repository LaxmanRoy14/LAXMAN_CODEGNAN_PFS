Numpy
------
--> Numpy(Numerical Python) is the foundational open-source library for scientific computing in python, providing high-performance, N-dimentional array objects (ndarray)
--> This enables efficient numerical computing with linear algebra, and data manipulation, serving as the basis for tools like Tensorlow and Scipy

```python
import numpy as np
```

```python
arr = np.array([1, 2, 3])
print(arr)
print(arr - 1)
```
